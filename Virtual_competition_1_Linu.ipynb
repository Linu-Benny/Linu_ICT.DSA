**1.** Create a DataFrame `df` from this dictionary `data` which has the index `labels`.

````python
# import libraries

import numpy as np
import pandas as pd

data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
````

````python
# creating Dataframe and give index option to give labels as index

df=pd.DataFrame(data=data,index=labels)
df
````

**20.** Draw a plot between dep_delay and arr_delay

````python
# filling the missing values in both the dep_dealy and arr_delay columns with 0

ny = ny.fillna(0)
````

````python
# check whether the values are filled or not

ny.info()
````

````python
# import matplotlib
import matplotlib.pyplot as plt

# here the columns dep_delay and arr_delay are numerical columns  so the scatter plot is the best option.

plt.figure(figsize=(8,7))
plt.scatter(ny.dep_delay,ny.arr_delay,color='green',edgecolor='y')

plt.title('Deperture Delay Vs Arrival Delay',color='darkgreen',fontweight='bold',fontsize=18)
plt.xlabel('Departure Delay',color='seagreen')
plt.ylabel('Arrival Delay',color='seagreen')

plt.xticks(fontsize=8,color='olive')
plt.yticks(fontsize=8,color='olive');
````

````python
# Plot using seaborn

import seaborn as sns

plt.figure(figsize=(8,7))
sns.scatterplot(data=ny,x='dep_delay',y='arr_delay',color='navy',edgecolor='r')
plt.title("Departure Delay Vs Arrival delay",color='navy',fontweight='bold',fontsize=16);
````
