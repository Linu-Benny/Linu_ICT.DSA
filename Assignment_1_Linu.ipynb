**5.Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.**

```python
# Importing required libraries

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1.Bar Graph

plt.figure(figsize=(5,5))
plt.bar(iris_data.Classification,iris_data.PL,color='lightcoral')

plt.xlabel('Classification',fontsize=10,color='darkred')
plt.ylabel('Petal Length (in cm)',fontsize=10,color='darkred')
plt.title('Mean of Petal Length by Classification',fontweight='bold',color='firebrick')

plt.xticks(fontsize=9)
plt.yticks(fontsize=9);         # Use semicolon instead of plt.show()
```

```python
# 2. Boxplot

sns.boxplot(data=iris_data,x=iris_data.Classification,y=iris_data.SW,hue=iris_data.Classification)

plt.ylabel("Sepal Width (in cm)",color='navy')
plt.xlabel('Classification',color='navy')
plt.title('Box Plot of Sepal Width By Clasification',color='forestgreen');

plt.xticks(fontsize=9)
plt.yticks(fontsize=9);
```

```python
# 3.Line Plot

plt.figure(figsize=(8, 6))
for species in iris_data['Classification'].unique():
    classify= iris_data[iris_data['Classification'] == species]
    plt.plot(classify['SL'], label=species, marker='^', linestyle='-')

plt.xlabel('Observations', color='midnightblue')
plt.ylabel('Sepal Length (cm)',color='midnightblue')
plt.legend(title='Classification')
plt.title('Sepal Length Trends by Classification',color='darkblue',fontweight='bold')

plt.xticks(fontsize=9)
plt.yticks(fontsize=9);
```

```python
# 4.Count Plot

iris_data.PW.value_counts().plot(kind='bar',color='tan')

plt.xticks(fontsize=8,color='peru')
plt.yticks(fontsize=8,color='peru')
plt.xlabel('Petal Width', color='maroon',fontsize=11)
plt.ylabel('Count',color='maroon',fontsize=11)
plt.title('Countplot for Petal Width',color='chocolate',fontweight='bold',fontsize=16);
```

```python
# 5. Histogram for SL SW PL & PW vs Count

plt.subplot(2, 2, 1)
sns.histplot(iris_data['SL'], kde=True)
plt.title("Sepal Length Distribution")
plt.tight_layout()

plt.subplot(2, 2, 2)
sns.histplot(iris_data['SW'], kde=True)
plt.title("Sepal Width Distribution")
plt.tight_layout()

plt.subplot(2, 2, 3)
sns.histplot(iris_data['PL'], kde=True)
plt.title("Petal Length Distribution")
plt.tight_layout()

plt.subplot(2, 2, 4)                                        # 'kde=True' adds a Kernel Density Estimate (KDE) plot, provide a smooth representation
sns.histplot(iris_data['PW'], kde=True)                     #      of the underlying probability density function along with the histogram.
plt.title("Petal Width Distribution")
plt.tight_layout()                                          # 'tight_layout' is used to avoid overlapping of subplots

plt.suptitle('Distribution of SL, SW, PL & PW',fontsize=16) # 'supplot' is used to give the main heading
plt.subplots_adjust(top=0.8)                                # 'subplot_adjust' is used to avoid overlapping of maintitle and subtitle
```
