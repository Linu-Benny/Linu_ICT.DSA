**To check whether there is any increase in sales after stepping into digital marketing**

Here we use ttest, because the size of the sample is less than 30 and population variance is unknown.

And we want to compare sales before and after stepping into digital marketing across all months, so paired ttest is used (ttest_rel).

Null Hypothesis (H0): There is no increase in sales after stepping into digital marketing.

Alternative Hypothesis (H1): There is an increase in sales after stepping into digital marketing.


```python
# Define the Hypothesis

Null="There is no increase in sales after stepping into digital marketing."
Alt="There is an increase in sales after stepping into digital marketing."
```

```python
# Importing ttest_rel from scipy
from scipy.stats import ttest_rel
```

```python
# Compute t statistic and p value
t_stat,p_value = ttest_rel(data['Sales_before_digital_add(in $)'],data['Sales_After_digital_add(in $)'])
```

```python
print(t_stat,p_value)
```

```python
# Print result of Hypothesis testing
# significance level is 5%.

if p_value<0.05 :
  print('Reject Null Hypothesis', '\n', 'ie.,', Alt)
else:
  print('Accept Null Hypothesis', '\n', Null)
```

**To check whether there is any dependency between the features “Region” and “Manager”.**

```python
# Chi-Square test of independence is used to check the depedency between the features.

from scipy.stats import chi2_contingency
```

Null Hypothesis (H0): There is no dependency between the "Region" and "Manager" features.

Alternative Hypothesis (H1): There is a dependency between the "Region" and "Manager" features.

```python
# Define the Hypothesis

H0="There is no dependency between the Region and Manager features."
H1="There is a dependency between the Region and Manager features."
```

```python
#The Dependency Analysis between the variables region and manager

table_data=pd.crosstab([data['Region']],[data['Manager']])
table_data
```

```python
#Chi square test of independence

chi_stat,p_value,dof,expected=chi2_contingency(table_data)
```

```python
print( "chi-Square statistic - ",chi_stat,'\n',"P Value - ",p_value,'\n','Degrees of freedom : ',dof,'\n','Expected Freq','\n',expected)
```

```python
# Print the results

if p_value<0.05:
  print("Reject Null Hypothesis",'\n','ie.,',H1)
else:
  print("Accept Null Hypothesis",'\n','ie.,',H0)
```
